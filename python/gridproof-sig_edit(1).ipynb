```python
class manbot():
    def __init__(self, name, xpos, ypos, coop_stat):
        self.name = name
        self.xpos = xpos
        self.ypos = ypos
        self.coop_stat = coop_stat

    def __repr__(self):
        return f"manbot(name={self.name} xpos={self.xpos}, ypos={self.ypos}, coop_stat={self.coop_stat})"
        
```

```python
import random

class mover():
    def __init__(self, grid, mcols, mrows): #, grid, rows, cols)

        self.rows = mrows
        self.cols = mcols
        self.grid = grid
        

       
    
        self.n = 0
        self.b = 0
    
        self.rowlist = []
        self.collist = []
    
        while self.n <= mrows - 1:
    
            self.rowlist.append(self.n)
    
            self.n = self.n + 1



            

            
    
        while self.b <= mcols - 1:
    
            self.collist.append(self.b)
    
            self.b = self.b + 1

        self.gridx = random.choice(self.rowlist)
        self.gridy = random.choice(self.collist)

        print(grid[self.gridy][self.gridx][0].xpos)
        print(grid[self.gridy][self.gridx][0].ypos)

        chancelist = [1,2,3,4,5]
        movechoice = random.choice(chancelist)

        
        
        if movechoice == 1:
            if self.gridy>0:
                grid[self.gridy][self.gridx][0].ypos = grid[self.gridy][self.gridx][0].ypos - 1
                movedbot = grid[self.gridy][self.gridx].pop(0)
                grid[self.gridy - 1][self.gridx].append(movedbot)
            else:
                grid[self.gridy][self.gridx][0].ypos = grid[self.gridy][self.gridx][0].ypos

        
        
        elif movechoice == 3:
            if self.gridy < mrows - 1:
                grid[self.gridy][self.gridx][0].ypos = grid[self.gridy][self.gridx][0].ypos + 1
                movedbot = grid[self.gridy][self.gridx].pop(0)
                grid[self.gridy + 1][self.gridx].append(movedbot)
            else:
                grid[self.gridy][self.gridx][0].ypos = grid[self.gridy][self.gridx][0].ypos

        
        
        elif movechoice == 4:
            if self.gridx>0:
                grid[self.gridy][self.gridx][0].xpos = grid[self.gridy][self.gridx][0].xpos - 1
                movedbot = grid[self.gridy][self.gridx].pop(0)
                grid[self.gridy][self.gridx - 1].append(movedbot) 
            else:
                grid[self.gridy][self.gridx][0].xpos = grid[self.gridy][self.gridx][0].xpos

        
        
        elif movechoice == 2:
            if self.gridx < mcols - 1:
                grid[self.gridy][self.gridx][0].xpos = grid[self.gridy][self.gridx][0].xpos + 1
                movedbot = grid[self.gridy][self.gridx].pop(0)
                grid[self.gridy][self.gridx + 1].append(movedbot)
            else:
                grid[self.gridy][self.gridx][0].xpos = grid[self.gridy][self.gridx][0].xpos

        
        
        elif movechoice == 5:
            grid[self.gridy][self.gridx][0].ypos = grid[self.gridy][self.gridx][0].ypos
            grid[self.gridy][self.gridx][0].xpos = grid[self.gridy][self.gridx][0].xpos

        
        
        #print(grid[self.gridy][self.gridx])
    

    
```
