```python
class manbot():
    def __init__(self, name, xpos, ypos, coop_stat):
        self.name = name
        self.xpos = xpos
        self.ypos = ypos
        self.coop_stat = coop_stat

    def __repr__(self):
        return f"manbot(name={self.name} xpos={self.xpos}, ypos={self.ypos}, coop_stat={self.coop_stat})"
```

```python
import plotly.express as px
import random

class mapper():
    def __init__(self):
    


        self.cols = int(input('how many columns:'))
        self.rows = int(input('how many rows:'))
        
        clonenum = self.cols * self.rows
        
        self.grid = [[[] for _ in range (self.cols)] for _ in range (self.rows)]
        
        n=0
        y=0
        self.b = 1
        self.prob_string = [1,2,3]
        L1 = 1
        pop_list = [1,2,3,4,5]
        
        while n <= (self.cols-1) and y <= (self.rows-1):


            pop_choice = random.choice(pop_list)

            while L1 <= pop_choice:
        
                clonename = ('Bot'+ str(self.b))
                bot_name = clonename
                xpos = y
                ypos = n
                coop_stat = random.choice(self.prob_string)
                
                clonename = manbot(bot_name, ypos, xpos, coop_stat)
                my_manbot = manbot(bot_name, ypos, xpos, coop_stat)
                self.grid[y][n].append(clonename)
                
                self.b=self.b+1

                L1 = L1 + 1

            L1 = 1

                
            if n<(self.cols - 1):
                n=n+1
            elif n==(self.cols - 1):
                n=0
                y=y+1
        
            
        #gridxpos = self.grid[1][1][0].xpos
        #print(gridxpos)

    def plotter(self):

        count_grid = [[len(cell) for cell in row] for row in self.grid]
        
        figure = px.imshow(count_grid, labels = dict(x='X coord', y='Y coord', color = 'Bot count'), text_auto=True, aspect='auto', color_continuous_scale='PuBu', origin = 'lower')
        figure.update_yaxes(scaleanchor="x", scaleratio=1)
        figure.update_layout(
            title="Bot Population Density",
            width=700, 
            height=700, 
            coloraxis_showscale=True)
        figure.show()
        
        
```

```python
my_mapper = mapper()
my_mapper.plotter()
```
